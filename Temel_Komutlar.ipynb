## Değişkenler (Variables)

Değişkenler, programlarımızda verileri depolamak için kullandığımız kaplardır. Bir değişkene isim verir ve o isme bir değer atarız. Python'da değişken tanımlarken veri tipini belirtmenize gerek yoktur, Python bunu otomatik olarak algılar.

```python
# Bir değişkene değer atama
isim = "Ahmet"
yas = 30
boy = 1.75

# Değişkenlerin değerlerini ekrana yazdırma
print("İsim:", isim)
print("Yaş:", yas)
print("Boy:", boy)

# Değişkenlerin değerlerini değiştirme
yas = yas + 1
print("Yeni Yaş:", yas)
```

## Veri Tipleri (Data Types)

Python'da farklı türde verileri depolamak için çeşitli veri tipleri bulunur. En yaygın olanlardan bazıları şunlardır:

*   **int (Integer):** Tam sayılar (örneğin, -2, 0, 150)
*   **float (Floating-point):** Ondalıklı sayılar (örneğin, 3.14, -0.5)
*   **str (String):** Metinler (örneğin, "Merhaba", "Python")
*   **bool (Boolean):** Mantıksal değerler (True veya False)

```python
# Farklı veri tiplerine örnekler
sayi_tam = 10       # int
sayi_ondalikli = 5.9 # float
metin = "Python öğreniyorum" # str
dogru_mu = True     # bool

# Veri tiplerini kontrol etme
print(type(sayi_tam))
print(type(sayi_ondalikli))
print(type(metin))
print(type(dogru_mu))
```

## Giriş ve Çıkış (Input and Output)

Programlarımızın kullanıcılarla etkileşim kurması için giriş (input) alması ve çıkış (output) vermesi gerekir.

*   **`input()`:** Kullanıcıdan veri almak için kullanılır. Kullanıcının girdiği her şey varsayılan olarak string (metin) olarak kabul edilir.
*   **`print()`:** Ekrana çıktı vermek için kullanılır. Farklı veri tiplerini ve birden fazla değeri aynı anda yazdırabilirsiniz.

```python
# Kullanıcıdan isim alma
isim = input("Adınız nedir? ")

# Kullanıcıdan yaş alma ve tam sayıya çevirme
yas = int(input("Yaşınız kaç? "))

# Alınan bilgileri ekrana yazdırma
print("Merhaba,", isim)
print("Yaşınız:", yas)

# Birden fazla değeri print ile yazdırma
print("Ad:", isim, "| Yaş:", yas)
```

## Koşullu İfadeler (Conditional Statements)

Koşullu ifadeler, belirli bir koşulun doğru olup olmadığına göre farklı kod bloklarının çalışmasını sağlar.

*   **`if`:** Belirtilen koşul doğruysa çalışacak kod bloğunu tanımlar.
*   **`elif`:** Önceki `if` veya `elif` koşulları yanlışsa ve bu `elif` koşulu doğruysa çalışacak kod bloğunu tanımlar.
*   **`else`:** Yukarıdaki `if` ve tüm `elif` koşulları yanlışsa çalışacak kod bloğunu tanımlar.

```python
# Yaşa göre farklı mesajlar yazdırma
yas = 25

if yas < 18:
  print("Henüz reşit değilsiniz.")
elif yas >= 18 and yas < 65:
  print("Yetişkinsiniz.")
else:
  print("Emeklilik yaşındasınız.")

# Bir sayının pozitif, negatif veya sıfır olduğunu kontrol etme
sayi = -10

if sayi > 0:
  print("Sayı pozitif.")
elif sayi < 0:
  print("Sayı negatif.")
else:
  print("Sayı sıfır.")
```

## Döngüler (Loops)

Döngüler, belirli bir kod bloğunu birden çok kez çalıştırmak için kullanılır.

*   **`for` döngüsü:** Bir dizinin (liste, string vb.) elemanları üzerinde gezinmek veya belirli sayıda tekrar yapmak için kullanılır.

```python
# Bir liste üzerindeki elemanları yazdırma
meyveler = ["elma", "muz", "kiraz"]
for meyve in meyveler:
  print(meyve)

# Belirli sayıda tekrar yapma (range() fonksiyonu ile)
for i in range(5): # 0'dan 4'e kadar sayar
  print("Tekrar:", i)

# Bir string üzerindeki karakterleri yazdırma
kelime = "Python"
for harf in kelime:
  print(harf)
```

*   **`while` döngüsü:** Belirtilen koşul doğru olduğu sürece çalışmaya devam eder. Koşul yanlış olduğunda döngü sona erer. Dikkatli kullanılmazsa sonsuz döngüye girebilir.

```python
# Belirli bir koşul doğru olduğu sürece tekrar yapma
sayac = 0
while sayac < 5:
  print("Sayaç:", sayac)
  sayac = sayac + 1 # veya sayac += 1

# Kullanıcı "çıkış" yazana kadar input almaya devam etme
giris = ""
while giris != "çıkış":
  giris = input("Çıkmak için 'çıkış' yazın: ")
  print("Siz şunu yazdınız:", giris)
```

## Fonksiyonlar (Functions)


Fonksiyonlar, belirli bir işi yapmak için bir araya getirilmiş kod bloklarıdır. Bir fonksiyon tanımladığınızda, o kod bloğunu daha sonra programınızın farklı yerlerinde tekrar tekrar kullanabilirsiniz. Bu, kodunuzu daha düzenli, okunabilir ve yönetilebilir hale getirir.

###Fonksiyon Tanımlama:

Bir fonksiyon tanımlamak için def anahtar kelimesi kullanılır, ardından fonksiyonun adı, parantezler () ve iki nokta üst üste : gelir. Fonksiyonun yapacağı işi belirten kod bloğu girintili olarak yazılır.

```python
def fonksiyon_adi():
  # Fonksiyonun yapacağı işler buraya yazılır
  print("Bu bir fonksiyon.")
```

###Fonksiyonu Çağırma:

Fonksiyonu tanımladıktan sonra, adını kullanarak çağırabilirsiniz.

```python
fonksiyon_adi() # Fonksiyonu çağırma
```

###Parametreler ve Argümanlar:

Fonksiyonlar, dışarıdan bilgi almak için parametreler kullanabilir. Parametreler, fonksiyon tanımındaki parantez içinde belirtilir. Fonksiyonu çağırırken bu parametrelere gönderdiğiniz değerlere argüman denir.

```python
def selamlama(isim): # 'isim' bir parametredir
  print("Merhaba,", isim)

selamlama("Ahmet") # "Ahmet" bir argümandır
selamlama("Ayşe")   # "Ayşe" başka bir argümandır
```

###Değer Döndürme:

Fonksiyonlar, bir işlem yaptıktan sonra bir değer döndürebilir. Bu, return anahtar kelimesi ile yapılır.

```python
def toplama(sayi1, sayi2):
  sonuc = sayi1 + sayi2
  return sonuc # sonucu döndürür

toplam = toplama(5, 3) # Fonksiyonun döndürdüğü değer 'toplam' değişkenine atanır
print("Toplam:", toplam)
```

###Varsayılan Parametre Değerleri:

Fonksiyon tanımlarken parametrelere varsayılan değerler atayabilirsiniz. Eğer fonksiyon çağrılırken bu parametreye bir değer gönderilmezse, varsayılan değer kullanılır.

```python
def selamlama(isim="Misafir"): # Varsayılan değer "Misafir"
  print("Merhaba,", isim)

selamlama()       # Çıktı: Merhaba, Misafir
selamlama("Veli") # Çıktı: Merhaba, Veli
```

## Veri Yapıları (Data Structures) - Listeler (Lists)

Listeler, birden çok öğeyi tek bir değişkende saklamak için kullanılır. Sıralıdırlar, değiştirilebilirler (mutable) ve farklı veri tiplerinden öğeler içerebilirler. Listeler köşeli parantez `[]` ile tanımlanır.

```python
# Boş bir liste oluşturma
bos_liste = []

# Elemanları olan bir liste oluşturma
meyveler = ["elma", "muz", "kiraz", "portakal"]

# Listeye eleman ekleme (listenin sonuna)
meyveler.append("üzüm")

# Belirli bir konuma eleman ekleme
meyveler.insert(1, "çilek") # 1. indekse "çilek" ekler

# Listeden eleman çıkarma (değerine göre)
meyveler.remove("muz")

# Listeden eleman çıkarma (indeksine göre)
silinen_meyve = meyveler.pop(0) # 0. indeksteki elemanı siler ve döndürür

# Listenin belirli bir indeksindeki elemana erişme
ilk_meyve = meyveler[0] # İlk eleman (indeks 0)
son_meyve = meyveler[-1] # Son eleman (negatif indeksleme)

# Listenin uzunluğunu alma
liste_uzunlugu = len(meyveler)

# Liste dilimleme (slicing)
ilk_iki_meyve = meyveler[0:2] # 0'dan 2'ye kadar (2 dahil değil)
geri_kalanlar = meyveler[2:] # 2'den sonuna kadar

print("Meyveler listesi:", meyveler)
print("Silinen meyve:", silinen_meyve)
print("İlk meyve:", ilk_meyve)
print("Son meyve:", son_meyve)
print("Liste uzunluğu:", liste_uzunlugu)
print("İlk iki meyve:", ilk_iki_meyve)
print("Geri kalanlar:", geri_kalanlar)
```

## Veri Yapıları (Data Structures) - Sözlükler (Dictionaries)

Sözlükler, anahtar-değer çiftleri halinde veri saklamak için kullanılır. Sırasızdırlar (Python 3.7 ve öncesi için), değiştirilebilirler ve anahtarlar benzersiz olmalıdır. Anahtarlar genellikle string veya sayı gibi değişmez (immutable) veri tiplerinden oluşur. Sözlükler süslü parantez `{}` ile tanımlanır.

```python
# Boş bir sözlük oluşturma
bos_sozluk = {}

# Elemanları olan bir sözlük oluşturma
ogrenci = {
    "isim": "Ahmet",
    "yas": 30,
    "bolum": "Bilgisayar Mühendisliği"
}

# Sözlüğe yeni eleman ekleme veya mevcut elemanı güncelleme
ogrenci["sehir"] = "Ankara" # Yeni eleman ekleme
ogrenci["yas"] = 31       # Mevcut elemanı güncelleme

# Sözlükten eleman silme (anahtarına göre)
del ogrenci["bolum"]

# Sözlüğün bir anahtarına karşılık gelen değere erişme
isim = ogrenci["isim"]
# Eğer anahtar yoksa hata verir, bu yüzden get() metodunu kullanmak daha güvenlidir
sehir = ogrenci.get("sehir", "Bilgi Yok") # Eğer 'sehir' yoksa "Bilgi Yok" döndürür

# Sözlüğün tüm anahtarlarını alma
anahtarlar = ogrenci.keys()

# Sözlüğün tüm değerlerini alma
degerler = ogrenci.values()

# Sözlüğün tüm anahtar-değer çiftlerini alma
elemanlar = ogrenci.items()

# Sözlük üzerinde döngü yapma (varsayılan olarak anahtarlar üzerinde döner)
print("Öğrenci Bilgileri:")
for anahtar in ogrenci:
  print(f"{anahtar}: {ogrenci[anahtar]}")

print("\nGüncellenmiş Öğrenci Sözlüğü:", ogrenci)
print("İsim:", isim)
print("Şehir:", sehir)
print("Anahtarlar:", list(anahtarlar)) # keys() bir view döndürür, listeye çevirebilirsiniz
print("Değerler:", list(degerler))   # values() bir view döndürür, listeye çevirebilirsiniz
print("Elemanlar:", list(elemanlar)) # items() bir view döndürür, listeye çevirebilirsiniz
```

## String İşlemleri (String Operations)

Stringler, Python'daki metinsel veri tipleridir. Tek tırnak (`'...'`) veya çift tırnak (`"..."`) içinde tanımlanırlar. Stringler üzerinde çeşitli işlemler yapabiliriz.

```python
# String oluşturma
greeting = "Merhaba"
name = 'Python'
cümle = "Python öğrenmek çok eğlenceli!"

# Stringleri birleştirme (concatenation)
tam_mesaj = greeting + " " + name
print(tam_mesaj)

# String tekrarı
yıldızlar = "*" * 10
print(yıldızlar)

# String uzunluğunu alma
cümle_uzunlugu = len(cümle)
print("Cümlenin uzunluğu:", cümle_uzunlugu)

# String içindeki karakterlere erişme (indeksleme)
ilk_harf = greeting[0]
son_harf = name[-1]
print("İlk harf:", ilk_harf)
print("Son harf:", son_harf)

# String dilimleme (slicing)
ilk_üç_harf = greeting[0:3] # 0'dan 3'e kadar (3 dahil değil)
son_dört_harf = name[2:]    # 2'den sonuna kadar
print("İlk üç harf:", ilk_üç_harf)
print("Son dört harf:", son_dört_harf)

# String metodları (bazı yaygın olanlar)
buyuk_harf = cümle.upper() # Tüm harfleri büyük yapar
kucuk_harf = cümle.lower() # Tüm harfleri küçük yapar
baslik_formatı = cümle.title() # Her kelimenin ilk harfini büyük yapar
kelimeler = cümle.split() # Boşluklara göre böler ve liste döndürür
degistirilmis_cümle = cümle.replace("eğlenceli", "harika") # Bir alt stringi diğeriyle değiştirir

print("Büyük harf:", buyuk_harf)
print("Küçük harf:", kucuk_harf)
print("Başlık formatı:", baslik_formatı)
print("Kelimeler:", kelimeler)
print("Değiştirilmiş cümle:", degistirilmis_cümle)

# String içinde arama yapma
python_var_mı = "Python" in cümle
print("'Python' kelimesi cümlede var mı?:", python_var_mı)

# String formatlama (f-string kullanarak)
sayi1 = 10
sayi2 = 20
toplam = sayi1 + sayi2
formatli_mesaj = f"'{sayi1}' ve '{sayi2}' sayılarının toplamı {toplam}'dir."
print(formatli_mesaj)
```

## Modüller ve İçeri Aktarma (Modules and Importing)

Python, farklı görevler için önceden yazılmış kodları içeren modüllere sahiptir. Bu modülleri programımıza "içeri aktararak" (importing) içindeki fonksiyonları, sınıfları ve değişkenleri kullanabiliriz. Bu, kod yazarken zaman kazandırır ve başkalarının yazdığı kullanışlı araçlardan faydalanmamızı sağlar.

*   **`import`:** Belirli bir modülün tamamını içeri aktarmak için kullanılır.
*   **`from ... import ...`:** Bir modülün içinden belirli bir fonksiyonu, sınıfı veya değişkeni içeri aktarmak için kullanılır.
*   **`import ... as ...`:** Bir modülü içeri aktarırken ona farklı bir isim (takma ad) vermek için kullanılır. Bu, özellikle modül adları uzunsa veya isim çakışmalarını önlemek için faydalıdır.

```python
# math modülünün tamamını içeri aktarma
import math

# math modülündeki sqrt() fonksiyonunu kullanma
karekok = math.sqrt(25)
print("25'in karekökü:", karekok)

# math modülündeki pi değişkenini kullanma
pi_degeri = math.pi
print("Pi değeri:", pi_degeri)

# random modülünden sadece randint() fonksiyonunu içeri aktarma
from random import randint

# randint() fonksiyonunu kullanma (artık math.randint() yazmaya gerek yok)
rastgele_sayi = randint(1, 10) # 1 ile 10 arasında rastgele tam sayı
print("Rastgele sayı (1-10):", rastgele_sayi)

# pandas modülünü pd olarak içeri aktarma (veri analizi için sık kullanılır)
import pandas as pd

# pandas DataFrame oluşturma (örnek kullanım)
data = {'isim': ['Ahmet', 'Ayşe', 'Mehmet'],
        'Yaş': [30, 25, 35]}
df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)
```
